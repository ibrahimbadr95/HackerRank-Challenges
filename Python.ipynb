## 6. Write a function


```python
def is_leap(year):
    leap = False    
    if year % 4 == 0 and year % 100 != 0:
        leap = True
    elif year % 100 == 0 and year % 400 == 0:
        leap = True
    return leap

year = int(input())
print(is_leap(year))
```

## 10. sWAP cASE


```python
def swap_case(s):
    result = ""
    for l in s:
        if l.isupper():
            l = l.lower()
        else:
            l = l.upper()
        result += l
    return result

if __name__ == '__main__':
    s = input()
    result = swap_case(s)
    print(result)
```

## 11. Nested Lists


```python
if __name__ == '__main__':
    record = []
    for _ in range(int(input())):
        name = input()
        score = float(input())
        record.append([name, score])

    scores = set()
    for s in record:
        scores.add(s[1])

    sorted_scores = sorted(scores)

    result = []
    for stu in record:
        if stu[1] == sorted_scores[1]:
            result.append(stu[0])
            # print(stu)
    result.sort()

    for i in result:
        print(i)
```

## 12. Finding the percentage


```python
def get_average_mark(student_marks, query_name):
    marks = student_marks[query_name]
    return sum(marks) / len(marks)

if __name__ == '__main__':
    n = int(input())
    student_marks = {}
    for _ in range(n):
        name, *line = input().split()
        scores = list(map(float, line))
        student_marks[name] = scores
        
    query_name = input()
    
    avg = get_average_mark(student_marks, query_name)
    
    print(f"{avg:.2f}")
```

## 14. itertools.permutations()


```python
# Enter your code here. Read input from STDIN. Print output to STDOUT
from itertools import permutations

inp = input().split()
sort = ''.join(sorted(inp[0], key=lambda x: (x.isdigit(), x)))

pre_list = list(permutations(sort, int(inp[1])))

for i in range(len(pre_list)):
    print(''.join(pre_list[i]))
```

## 16. Introduction to Sets


```python
def average(array):
    input_set = set(array)
    return sum(input_set) / len(input_set)

if __name__ == '__main__':
    n = int(input())
    arr = list(map(int, input().split()))
    result = average(arr)
    print(result)
```
