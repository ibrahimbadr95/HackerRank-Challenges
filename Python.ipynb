## 6. Write a function


```python
def is_leap(year):
    leap = False    
    if year % 4 == 0 and year % 100 != 0:
        leap = True
    elif year % 100 == 0 and year % 400 == 0:
        leap = True
    return leap

year = int(input())
print(is_leap(year))
```

## 8. List Comprehensions


```python
if __name__ == '__main__':
    x = int(input())
    y = int(input())
    z = int(input())
    n = int(input())
    
    result = [[a, b, c] for a in range(x+1) for b in range(y+1) for c in range(z+1) if a+b+c != n]
    print(result)
```

## 10. sWAP cASE


```python
def swap_case(s):
    result = ""
    for l in s:
        if l.isupper():
            l = l.lower()
        else:
            l = l.upper()
        result += l
    return result

if __name__ == '__main__':
    s = input()
    result = swap_case(s)
    print(result)
```
