## 2. Compare the Triplets 


```python
#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'compareTriplets' function below.
#
# The function is expected to return an INTEGER_ARRAY.
# The function accepts following parameters:
#  1. INTEGER_ARRAY a
#  2. INTEGER_ARRAY b
#

def compareTriplets(a, b):
    result = [0, 0]
    for (a_el, b_el) in zip(a, b):
        if a_el > b_el:
            result = [result[0] + 1, result[1]]
        elif b_el > a_el:
            result = [result[0], result[1] + 1]
        else:
            result = result

    return result

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    a = list(map(int, input().rstrip().split()))

    b = list(map(int, input().rstrip().split()))

    result = compareTriplets(a, b)

    fptr.write(' '.join(map(str, result)))
    fptr.write('\n')

    fptr.close()
```

## 3. Diagonal Difference


```python
#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'diagonalDifference' function below.
#
# The function is expected to return an INTEGER.
# The function accepts 2D_INTEGER_ARRAY arr as parameter.
#

def diagonalDifference(arr):
    main_d = []
    sec_d = []
    for i in range(len(arr)):
        for j in range(len(arr)):
            if i == j: 
                main_d.append(arr[i][j])

    for i in range(len(arr)):
        for j in range(len(arr)):
            if j == (len(arr) - i -1):
                sec_d.append(arr[i][j])
    return abs(sum(main_d) - sum(sec_d))


if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    n = int(input().strip())

    arr = []

    for _ in range(n):
        arr.append(list(map(int, input().rstrip().split())))

    result = diagonalDifference(arr)

    fptr.write(str(result) + '\n')

    fptr.close()
```

## 5. Mini-Max Sum


```python
import math
import os
import random
import re
import sys

#
# Complete the 'miniMaxSum' function below.
#
# The function accepts INTEGER_ARRAY arr as parameter.
#

def miniMaxSum(arr):
    min = arr[0]
    max = arr[0]
    
    for elm in arr:
        if elm > max:
            max = elm
    
    for elm in arr:
        if elm < min:
            min = elm
    print(f"{sum(arr) - max} {sum(arr) - min}")

if __name__ == '__main__':

    arr = list(map(int, input().rstrip().split()))

    miniMaxSum(arr)
```

## 6. Time Conversion


```python
import math
import os
import random
import re
import sys

#
# Complete the 'timeConversion' function below.
#
# The function is expected to return a STRING.
# The function accepts STRING s as parameter.
#
def timeConversion(s):
    tokens = s.split(':')
    time = tokens[2][2]
    sec = re.match(r'(\d+)([A-Za-z]+)', tokens[2])
    result = []
     
    if time == 'P':
        if (int(tokens[0]) == 12):
            hour = str(int(tokens[0]))
        else:
            hour = str(int(tokens[0]) + 12)
        result.append(hour)
        result.append(tokens[1])
        result.append(sec.group(1))
    elif time == 'A' and tokens[0] == "12":
        result.append(str(int(tokens[0]) - 12)+ '0')
        result.append(tokens[1])
        result.append(sec.group(1))
 
    else:
        result.append(tokens[0])
        result.append(tokens[1])
        result.append(sec.group(1))
                 
    return ':'.join(result)

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    s = input()

    result = timeConversion(s)

    fptr.write(result + '\n')

    fptr.close()
```

## 7. Counting Sort 1


```python
import math
import os
import random
import re
import sys

#
# Complete the 'countingSort' function below.
#
# The function is expected to return an INTEGER_ARRAY.
# The function accepts INTEGER_ARRAY arr as parameter.
#

def countingSort(arr):
    result = [0] * 100
    for i in range(len(arr)):
        result[arr[i]] = result[arr[i]] + 1
    
    return result

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    n = int(input().strip())

    arr = list(map(int, input().rstrip().split()))

    result = countingSort(arr)

    fptr.write(' '.join(map(str, result)))
    fptr.write('\n')

    fptr.close()
```

## 8. Flipping the Matrix


```python
#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'flippingMatrix' function below.
#
# The function is expected to return an INTEGER.
# The function accepts 2D_INTEGER_ARRAY matrix as parameter.
#

def flippingMatrix(matrix):
    n = len(matrix)
    result = 0 
    
    for i in range(int(n/2)):
        for j in range(int(n/2)):
            result += max(matrix[i][j], matrix[i][n-j-1], matrix[n-i-1][j], matrix[n -i - 1][n - j - 1])
    return result

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    q = int(input().strip())

    for q_itr in range(q):
        n = int(input().strip())

        matrix = []

        for _ in range(2 * n):
            matrix.append(list(map(int, input().rstrip().split())))

        result = flippingMatrix(matrix)

        fptr.write(str(result) + '\n')

    fptr.close()
```

## 9. Lonely Integer


```python
import math
import os
import random
import re
import sys

#
# Complete the 'lonelyinteger' function below.
#
# The function is expected to return an INTEGER.
# The function accepts INTEGER_ARRAY a as parameter.
#

def lonelyinteger(a):
    result = a[0]
    for elm in a:
        if a.count(elm) == 1:
            result = elm
            return result


if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    n = int(input().strip())

    a = list(map(int, input().rstrip().split()))

    result = lonelyinteger(a)

    fptr.write(str(result) + '\n')

    fptr.close()
```
